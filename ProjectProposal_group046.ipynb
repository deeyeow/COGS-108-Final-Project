## Research Question
&emsp;Considering the amount of resources spent on medical research and its healthcare system, how well was the USA prepared for Covid19 in contrast to other countries?


## Background and Prior Work
&emsp;COVID-19 has wreaked havoc on the entire world. It has clawed its way into our personal lives and cut short the lives of many loved ones, threatening the very spirit of humanity. As students living in the US, we could not help but wonder if our country could have handled the pandemic any differently. While the US invests a lot of money into medical research, it was not immune to the early onset of ventilator and mask shortages. Mortality rates remained higher for longer, relative to many other countries. Safe to say, the US was not prepared for COVID-19.
	
&emsp;Armed with this knowledge, we wish to compare the US’s approach to COVID-19 with other countries. Switzerland and Germany (known for their quality healthcare systems) will make great case studies, as will Brazil (significant rise in cases) and Saudi Arabia (free healthcare, currently in recovery). Hopefully, we can uncover the factors that hindered our pandemic-readiness, and push for improvement before disaster strikes again.

## Hypothesis
&emsp;The United States spends a large amount of money on healthcare and medical research every year. However, we have all witnessed that the USA’s healthcare system was not as prepared for COVID-19 as other countries were (consider Switzerland and German). We hypothesize that a country’s effectiveness in handling an emergency like the Coronavirus depends on a variety of other factors, namely its governmental decision-making and its available resources.
Data:

&emsp;The ideal datasets for this project would correlate Covid-19 data (cases, death rates) by country, to other metrics that we wish to investigate: the amount of money invested in the healthcare system, education system, number of hospitals, physicians, and pharmaceutical companies in the country, etc. 
 
&emsp;Ideally, the datasets would show a detailed timeline of the number of cases,  number of hospitalized and recovered patients, amount of testing, and mortality rates in a country. With a detailed timeline, we can compare the reaction time of different governments. To fairly contrast countries’ responses, we will first streamline and quantify the definitions of our chosen observations and qualitative standards.
 
&emsp;Because we care about the money invested in the healthcare system, we want data that displays the amount of money each country spent in reducing the spread of the coronavirus. We will consider how efficiently laws were implemented (speed of government and compliance of the citizens). We also need a baseline. We want to look at each country’s economic state and the state of healthcare system before and during Covid-19.
 
&emsp;We will collect data from the official healthcare administration of each country and the World Health Organization. Data pulled from these resources will be clean, organized, comprehensive, accurate, and already follows the guidelines of ethical data collection. The datasets should be consistent enough that merges will smooth and will allow appropriate comparisons. Ideally, the data will be stored in CSV files, easily accessible through the internet.
